## Task 1:
Consider the following experiment:

1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_i,i = 1,2,...,k$ and remains the same from trial to trial. Notice that $p_1+p_2+p_3+···+p_k =1.$
4. The trials are independent.
5. The random variables of interest are $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

Question 1: Show that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i $ 

<br>

$$\textbf{Answer}$$

<br>

&bullet; To get expected value $E[y]$, we use the formula: $E[y] = \Sigma y \cdotp P(y)$

and adding that to the joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

 we have: $$E[y] = \Sigma y \cdotp \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$$


&bullet; Add the given $Y_1 +Y_2 +Y_3 +···+Y_k =n$ to the formula, and since trials are independent, we can rewrite the joint probability as:

$$E[y] = \sum (y_1 \cdot P_1 + y_2 \cdot P_2 + y_3 \cdot P_3 + \ldots + y_n \cdot P_n)$$
$$E[y] = \sum y_i \cdot P_i  $$

&bullet; Since $P_i = p_i$, we get: 

$$E[y] = \sum y_i \cdot p_i  $$
$$E[y] = n \cdot p_i  $$

$\textbf{Conclusion:}$ We have proven that the expected value of the random variable $y$ is $n \cdot p_i $

## **Task 2:** 
Consider the following experiment:

1. The experiment consists of a fixed number, $n$, of identical trials.
2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$ and remains the same from trial to trial. The probability of a failure is equal to q = $(1 − p)$.
4. The trials are independent.
5. The random variable of interest is $Y$ , the number of successes observed during the $n$ trials.

**Question 1:**
From the following steps shown above, derive the probability distribution of the experiment.

<br>

$$\textbf{Answer}$$

<br>


$$[ P(Y = k) = \binom{n}{k} p^k (1 - p)^{n - k} ]$$

where:

- $Y$ # of successes during $n$ trials.
- $k$ # of successes from 0 to $n$>
- $p$ probability of success on a single trial.
- $(1 - p)$ probability of failure on a single trial.
- $\binom{n}{k}$ number of combinations of $k$ successes out of $n$ trials.

**Question 2:**
Show that the expectation of this probability distribution is $E[y] = \Sigma y \cdotp P(y) = np$.

<br>

$$\textbf{Answer}$$

<br>

- using the definition of expectation for a discrete random variable $Y$.

$$E[Y] = \sum_{\text{all } y} y \cdot P(Y = y)$$

- Given the experiment described and the probability distribution derived from it, where $Y$ represents the number of successes observed during $n$ trials, we have:

$$P(Y=y) = \binom{n}{y} \cdot p^y \cdot (1 - p)^{n - y}$$

where:

- $n$ total number of trials,
- $y$ number of successes,
- $p$ probability of success on a single trial,
- $(1 - p)$ probability of failure on a single trial
- $\binom{n}{y}$ binomial coefficient.

Step 3: Compute and simplify

$$\begin{align*}
E[Y] &= \sum_{y=0}^{n} y \cdot P(Y=y) \\
&= \sum_{y=0}^{n} y \cdot \binom{n}{y} \cdot p^y \cdot (1-p)^{n-y}
\end{align*}$$



$$(p + (1 - p))^n = \sum_{y=0}^{n} \binom{n}{y} \cdot p^y \cdot (1-p)^{n-y}$$

we see that the sum $\sum_{y=0}^{n} y \cdot \binom{n}{y} \cdot p^y \cdot (1-p)^{n-y} \text{ is equivalent to } np.$

&#8756; $E[Y] = np$, which proves that the expectation of the probability distribution of $Y$ is $np$

## **Task 3:** 
Consider a probility distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.

**Question 1:**
Find a general formula for the expected value of this distribution.

$\textbf{Answers}$

- expression for expected value $E(y)$:
$$[E(y) = e^{-\lambda} \left( \lambda \sum_{k=0}^{\infty} \frac{\lambda^k}{k!} + 1 \right)$$

- sum $\sum_{k=0}^{\infty} \frac{\lambda^k}{k!}$ is the Taylor series expansion of $(e^\lambda)$, we then substitute it in: 
$$E(y) = e^{-\lambda} (\lambda e^\lambda + 1)$$

- Distribute $(e^{-\lambda})$ 
$$E(y) = \lambda e^{-\lambda} e^\lambda + e^{-\lambda} $$
$$E(y) = \lambda \cdot 1 + e^{-\lambda} $$
$$E(y) = \lambda + e^{-\lambda} $$

$\textbf{Conclusion: }$ $e^{-\lambda}$ is constant factor that doesnt depend on $y$, thus not affecting the expected value. &#8756; the expected value of the distribution is $\lambda$.

**Question 2:**
A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.

$\textbf{Answers}$

- Starting with $p(y)$ as the expression for probability distribution :

$$p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$$

$(parameters)$
- Mean density of seedlings $(\lambda)$ = 5 per square yard
- Number of sampling regions $(n)$ = 10

- $\text{Expected number of seedlings in one square yard} = \lambda = 5$

- Poisson distribution to find probability of no seedlings in one square yard:
$$P(Y = 0) = \frac{5^0}{0!} e^{-5} = e^{-5}$$

- Apply probability to all ten sampling regions, which are independent events:
$$P(\text{None of the regions contain seedlings}) = (e^{-5})^{10} = e^{-50}$$

$\textbf{Conclusion: } $chances of none of the ten "1 square yard" sampling regions will containt seedlings: $e^{-50}$.
