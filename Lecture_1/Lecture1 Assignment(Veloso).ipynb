1. Recall the definition of a rational number, denoted as $\mathbb{Q}$. Prove that the Euler's number $e = \Sigma_{k=0}^\infty \frac{1}{k!} \notin \mathbb{Q}$. A factorial is defined as $k! = (k)(k-1)(k-2)(k-3)..., \forall k \in \mathbb{Z}^+$, note that $0! = 1$. Furthermore, a sum notation $\Sigma_{k=0}^\infty k = 0+ 1 + 2 + 3 +....+...$

$\textbf{Answer}$

$\textbf{Proof By Contradiction:}$<br>

Step 1: To counter or contradict, assume that $e$ is rational. We can then express $e$ as a fraction $\frac{p}{q}$ where $p$ and $q$ are integers with no common factors other than 1, and $q \neq 0$.

Step 2: Recall the definition of $e$ as the sum of the infinite series:


$ e = \Sigma_{k=0}^\infty \frac{1}{k!}$
​
 
Step 3: Consider the partial sums of the series. Let $S_n$ indicate the $n$th partial sum:

$S_n = \Sigma_{k=0}^\infty \frac{1}{k!}$

Step 4: Know that $e$ is defined as the limit of these partial sums as $n$ approaches infinity:

$e = \lim_{n \to \infty} S_n$
 
Step 5: $S_n$ must approach a rational limit as $n$ goes to infinity, since $e$ is rational.

Step 6: We can also see that each term in the series $\frac{1}{k!}$ is positive. &#8756; each partial sum $S_n$ is increasing.

Step 7: Consider the factorial terms $k!$. For any $k \geq 1$, we have $k! > k$, which implies $\frac{1}{k!} < \frac{1}{k}$.

Step 8: We also notice the Harmonic series here since the series diverges; $\sum_{k=0}^\infty \frac{1}{k}$

Step 9: Since $\frac{1}{k!} < \frac{1}{k}$, we can conclude that the series $\sum_{k=0}^\infty \frac{1}{k!}$ converges.

Step 10: If $S_n$ converges to a rational number, it must be bounded.

Step 11: But convergence of $\sum_{k=0}^\infty \frac{1}{k!}$ contradicts the unboundedness of the harmonic series.

$\textbf{Conclusion}$: Since assuming $e$ is rational leads to a contradiction, we conclude that $e$ is not rational, i.e., $e \notin \mathbb{Q}$.   

$\textbf{2. Prove Minkowski's Inequality for sums, $\forall \ p>1, (a_k, b_k)>0$:}$
\begin{gather} 
\begin{bmatrix} 
\Sigma_{k=1}^n |a_k + b_k|^p 
\end{bmatrix}^\frac{1}{p}
\leq
\begin{bmatrix} 
\Sigma_{k=1}^n |a_k|^p 
\end{bmatrix}^ \frac{1}{p}
+
\begin{bmatrix}
\Sigma_{k=1}^n |b_k|^p
\end{bmatrix}^\frac{1}{p}
\end{gather}

$\text{Answer:}$

Step 1: Using Laws of Exponents:

- Let $n = $ any base number, then ${(n^p)}^\frac{1}{p}=n$
- For any base number $n$ we know that ${(n^p)}^\frac{1}{p}=n$, which implies that the $p$-th root of $n^p$ is $n$. Applying this to the sums of $|a_k|^p$ and $|b_k|^p$, we have:

\begin{gather} 
\Sigma_{k=1}^n |a_k + b_k|
\leq
\Sigma_{k=1}^n |a_k|
+
\Sigma_{k=1}^n |b_k|
\end{gather}

Step 2: Use of Property of Summation

\begin{gather} 
\Sigma_{k=1}^n |a_k + b_k|
\leq
\Sigma_{k=1}^n |a_k| + |b_k|
\end{gather}

Step 3: Prove by Triangle Inequality

- Disregarding the Sigma Notation we get:
\begin{gather}
|a_k + b_k|
\leq
|a_k| + b_k| 
\end{gather}

- Since, $(a_k, b_k) > 0$, then $(a + b) = a + b$, and $|a| = a_k$, $|b| = b$:
- Substitute:
\begin{gather}
|a + b| = a + b = |a| + |b|
\end{gather}

$\textbf{Conclusion}$: Thus, the inequality is proven.

$\textbf{3. Prove the triangle inequality $|x+y| \leq |x| + |y|, \forall (x,y) \in \mathbb{R}$}$

$\text{Answer:}$

Step 1: Definition of an absolute value

- Definition
\begin{gather}
{|x| = }
\begin{cases}
x,&{x \geq 0}\\ -x,&{x < 0}
\end{cases}
\end{gather}

- Thus, we must solve for 4 cases:

\begin{gather}
\text{Case 1: }{x \geq 0, y \geq 0} \\
\text{Case 2: }{x < 0, y < 0} \\
\text{Case 3: }{x \geq 0, y < 0} \\
\text{Case 4: }{x < 0, y \geq 0} \\
\end{gather}

Case 1: $x \geq 0, y \geq 0$

- Let, $|x + y| = x + y$, and $|x| = x, |y| = y$
- Substitution:
\begin{gather}
|x + y| = x + y = |x| + |y|
\end{gather}

Case 2: $x < 0, y < 0$

- Let, $|x + y| = -(x + y)$, and $|x| = -x, |y| = -y$
- Substitution:
\begin{gather}
|x + y| = -(x + y) = -x + (-y) = |x| + |y|
\end{gather}

Case 3: $x \geq 0, y < 0$

- Let, $|x| = x, |y| = -y$
- Substitution:
\begin{gather}
|x + y| \leq x - y
\end{gather}
- Since, $x$  is non-negative and $y$ is negative, $x - y$ could be either positive or negative. So, let's consider both possibilities:

For $x + y \geq 0$:
\begin{gather}
x + y \leq x - y \\  
y \leq -y 
\end{gather}

For $x + y < 0$:
\begin{gather}
-(x + y) \leq x - y \\
-x - y \leq x - y \\
-x \leq x
\end{gather}

Case 4: $x < 0, y \geq 0$

- Let, $|x| = -x, |y| = y$
- Substitution:
\begin{gather}
|x + y| \leq -x + y
\end{gather}
- Solve for sub-cases of $|x + y|$:
\begin{gather}
x + y, x + y \geq 0 \\
-(x + y), x + y < 0 \\
\end{gather}

For $x + y \geq 0$:
\begin{gather}
x + y \leq -x + y \\
x \leq -x
\end{gather}

For $x + y < 0$:
\begin{gather}
-(x + y) \leq -x + y \\
-x - y \leq -x + y \\
-y \leq y
\end{gather}

$\textbf{Conclusion}$: Therefore, in all cases, $|x+y| \leq |x| + |y| \in \mathbb{R} $

$\textbf{4. Prove Sedrakayan's Lemma $\forall u_i, v_i \in \mathbb{R}^+$:}$

\begin{gather}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i}
\leq
\Sigma_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}

$\text{Answer:}$

#### Solution:


Step 1: Sedrakayan’s Lemma can be proved directly by demonstrating that for all $(u_i, v_i \in \mathbb{R}^+)$, the inequality (below) is true

$$\left(\frac{\sum_{i=1}^n u_i}{\sum_{i=1}^n v_i}\right)^2 \leq \frac{\sum_{i=1}^n (u_i)^2}{v_i}$$

Step 2: The <u>Cauchy-Schwarz Inequality</u> states that for any real numbers $(a_i)$ and $(b_i)$, 

$$\left(\sum_{i=1}^n a_i b_i\right)^2 \leq \left(\sum_{i=1}^n (a_i)^2\right) \left(\sum_{i=1}^n (b_i)^2\right)$$

Step 3: Utilizing the Cauchy-Shwarz Inequality, Let $a_i = \sqrt{u_i}$ and $b_i = \frac{1}{\sqrt{v_i}}$. Then,

$$\left(\sum_{i=1}^n \sqrt{u_i} \cdot \frac{1}{\sqrt{v_i}}\right)^2 \leq \left(\sum_{i=1}^n u_i\right) \left(\sum_{i=1}^n \frac{1}{v_i}\right)$$

#### Simplify:

Step 4: **LH** =

$$\left(\frac{\sum_{i=1}^n u_i}{\sum_{i=1}^n v_i}\right)^2$$

Step 5: **RH** =

$$\frac{\sum_{i=1}^n (u_i)^2}{v_i}$$
 &#8756;
since LH &le; RH, we have proven Sedrakayan’s Lemma..
