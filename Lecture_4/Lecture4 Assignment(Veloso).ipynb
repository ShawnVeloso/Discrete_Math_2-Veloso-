```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph
G = nx.Graph()
plt.figure(figsize=(10, 8))

# Add nodes
G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G"])

# Add edges with weights
edgelist = [
    ("A", "C", 10), ("A", "I", 5), ("C", "B", 20), ("C", "E", 15),
    ("C", "D", 30), ("B", "E", 25), ("I", "E", 40), ("E", "F", 35),
    ("F", "G", 50), ("G", "E", 45)
]
G.add_weighted_edges_from(edgelist)

# Define layout
pos = nx.planar_layout(G)  # Seed for reproducible layout

# Draw edges with labels
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={(u, v): w for u, v, w in edgelist},
    font_color='red'
)

# Draw the original graph
nx.draw(
    G, pos, edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    with_labels=True
)

# Find the minimum spanning tree using Kruskal's algorithm
mst = nx.minimum_spanning_tree(G)

# Draw the MST
nx.draw_networkx_edges(mst, pos, edge_color='coral', width=2)

plt.show()
```

$$\textbf{DIJKSTRA'S ALGORITHM}$$

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph
G = nx.Graph()
plt.figure(figsize=(10, 8))

# Add nodes
G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G", "I"])

# Add edges with weights
edgelist = [
    ("A", "C", 10), ("A", "I", 5), ("C", "B", 20), ("C", "E", 15),
    ("C", "D", 30), ("B", "E", 25), ("I", "E", 40), ("E", "F", 35),
    ("F", "G", 50), ("G", "E", 45)
]
G.add_weighted_edges_from(edgelist)

# Define layout
pos = nx.planar_layout(G)

# Draw edges with labels
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={(u, v): w for u, v, w in edgelist},
    font_color='red'
)

# Draw the original graph
nx.draw(
    G, pos, edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    with_labels=True
)

# Find the shortest paths using Dijkstra's algorithm from node A
shortest_paths = nx.single_source_dijkstra_path(G, source="A", weight="weight")

# Highlight the shortest path to node G by coloring the corresponding edges in red
shortest_path_to_G = shortest_paths["G"]
for i in range(len(shortest_path_to_G) - 1):
    nx.draw_networkx_edges(G, pos, edgelist=[(shortest_path_to_G[i], shortest_path_to_G[i+1])], edge_color='cornflowerblue', width=2)

# Print the shortest path to node G
shortest_path_length_to_G = nx.shortest_path_length(G, source='A', target='G', weight='weight')
print(f"To node G: {' -> '.join(shortest_path_to_G)} (length {shortest_path_length_to_G})")

plt.show()
```
